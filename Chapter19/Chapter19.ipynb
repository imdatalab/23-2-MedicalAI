##예제 1: CartPole-v1
CartPole 게임에서 막대기를 균형있게 유지하는 문제를 해결합니다.

```python
import gym
import numpy as np
import matplotlib.pyplot as plt
from IPython import display as ipythondisplay

# 환경 생성
env = gym.make('CartPole-v1') # gym.make란, OpenAI Gym 라이브러리에서 제공하는 함수 중 하나로 Gym환경 생성 시 사용
state = env.reset() # 강화학습 시작점

# 간단한 정책 정의: 막대기가 기울어지는 방향으로 카트를 움직입니다.
def simple_policy(state): # CartPole 환경의 상태는 [카트위치, 카트 속도, 막대기 각도, 막대기 각속도]와 같은 정보로 포함
    angle = state[2]# 막대기 각도 값이 angle변수에 저장
    return 0 if angle < 0 else 1

# 각 에피소드를 실행하는 루프
for _ in range(1000):
    plt.imshow(env.render(mode='rgb_array'))
    ipythondisplay.clear_output(wait=True)
    ipythondisplay.display(plt.gcf())

    action = simple_policy(state)
    state, _, done, _ = env.step(action)

    if done:
        break

env.close()
ipythondisplay.clear_output(wait=True)
'''
(wait=False)로 지정할 경우,
사용하면 이전 출력을 즉시 지우고 새로운 출력을 업데이트하므로
출력이 실시간으로 업데이트됨. 따라서 환경 실행 중에 상호작용적으로 시각화를 업데이트할 수 있습니다~
'''

```

## 예제 2: MountainCar-v0
Mountain Car 게임에서 차를 산 정상에 올리는 문제를 해결합니다.

```python
import gym
import matplotlib.pyplot as plt
from IPython import display as ipythondisplay

# 환경 생성
env = gym.make('MountainCar-v0')
state = env.reset()

# 간단한 정책 정의: 속도에 따라 액셀러레이터를 조절합니다.
def simple_policy(state):
    position, velocity = state #MountainCar의 환경상태는 위치와 속도 정보를 포함
    return 2 if velocity > 0 else 0

# 각 에피소드를 실행하는 루프
for _ in range(1000):
    plt.imshow(env.render(mode='rgb_array'))
    ipythondisplay.clear_output(wait=True)
    ipythondisplay.display(plt.gcf())

    action = simple_policy(state)
    state, _, done, _ = env.step(action)

    if done:
        break

env.close()
ipythondisplay.clear_output(wait=True)
```

## 예제 3: Acrobot-v1
Acrobot 게임에서 로봇 팔을 목표 위치까지 움직이는 문제를 해결

```python
import gym
import matplotlib.pyplot as plt
from IPython import display as ipythondisplay

# 환경 생성
env = gym.make('Acrobot-v1')
state = env.reset()

# 간단한 정책 정의: 무작위로 액션을 선택합니다.
def simple_policy(state):
    return env.action_space.sample()

# 각 에피소드를 실행하는 루프
for _ in range(1000):
    plt.imshow(env.render(mode='rgb_array'))
    ipythondisplay.clear_output(wait=True)
    ipythondisplay.display(plt.gcf())

    action = simple_policy(state)
    state, _, done, _ = env.step(action)

    if done:
        break

env.close()
ipythondisplay.clear_output(wait=True)
```
